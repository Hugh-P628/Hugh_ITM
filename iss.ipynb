```python
import matplotlib.pyplot as plt
import csv

# Load ISS data
latitudes = []
longitudes = []

with open("iss_tracking_data.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        latitudes.append(float(row["latitude"]))
        longitudes.append(float(row["longitude"]))

# Plot the path
plt.figure(figsize=(12, 6))
plt.plot(longitudes, latitudes, marker='o', linestyle='-', color='blue', markersize=2)
plt.title("ISS Path (Lat vs Long)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.xlim(-180, 180)
plt.ylim(-90, 90)

# Optional: mark start and end points
plt.scatter(longitudes[0], latitudes[0], color='green', label='Start')
plt.scatter(longitudes[-1], latitudes[-1], color='red', label='End')
plt.legend()

plt.tight_layout()
plt.show()
```
